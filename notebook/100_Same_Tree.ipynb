Definition for a binary tree node.<br>
class TreeNode(object):<br>
    def __init__(self, x):<br>
        self.val = x<br>
        self.left = None<br>
        self.right = None

```python
class Solution(object):
    def isSameTree(self, p, q):
        """
        :type p: TreeNode
        :type q: TreeNode
        :rtype: bool
        """
        if p == q:
            return True
        try:
            left = right = True
            if p.val == q.val:
                left = self.isSameTree(p.left, q.left)
                right = self.isSameTree(p.right, q.right)
                return (left and right)
        except:
            return False
        return False
```
