    def findCircleNum(self, M):<br>
        # Union Find<br>
        union = Union()<br>
        for i in range(len(M)):<br>
            union.add(i)<br>
        for i in range(len(M)):<br>
            for j in range(i + 1, len(M)):<br>
                if M[i][j] == 1:<br>
                    union.union(i, j)<br>
        return union.count

class Union(object):<br>
#     
<br>
#     weighted quick union find<br>
    
<br>
    def __init__(self):<br>
        # both dic and list is fine<br>
        self.id = {}<br>
        self.sz = {}<br>
        self.count = 0

    def count(self):<br>
        return self.count

    def connected(self, p, q):<br>
        return self.find(p) == self.find(q)

    def add(self, p):<br>
        # init<br>
        self.id[p] = p<br>
        self.sz[p] = 1<br>
        self.count += 1

    def find(self, p):<br>
#         
<br>
#         find root of p, and compress path<br>
        
<br>
        while p != self.id[p]:<br>
            self.id[p] = self.id[self.id[p]]<br>
            p = self.id[p]<br>
        return p

    def union(self, p, q):<br>
#         
<br>
#         connect p and q<br>
        
<br>
        i, j = self.find(p), self.find(q)<br>
        if i == j:<br>
            return<br>
        if self.sz[i] > self.sz[j]:<br>
            i, j = j, i<br>
        self.id[i] = j<br>
        self.sz[j] += self.sz[i]<br>
        self.count -= 1
