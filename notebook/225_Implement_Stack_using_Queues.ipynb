class Stack(object):<br>
    def __init__(self):<br>
#         
<br>
#         initialize your data structure here.<br>
        
<br>
        self.queue1 = []<br>
        self.queue2 = []<br>
<br>
    def push(self, x):<br>
#         
<br>
#         :type x: int<br>
#         :rtype: nothing<br>
        
<br>
        self.queue1.append(x)<br>
<br>
    def pop(self):<br>
#         
<br>
#         :rtype: nothing<br>
        
<br>
        while len(self.queue1) > 1:<br>
            curr = self.queue1.pop(0)<br>
            self.queue2.append(curr)<br>
        if len(self.queue1) == 1:<br>
            self.queue1.pop(0)<br>
        while len(self.queue2):<br>
            curr = self.queue2.pop(0)<br>
            self.queue1.append(curr)<br>
<br>
    def top(self):<br>
#         
<br>
#         :rtype: int<br>
        
<br>
        while len(self.queue1) > 1:<br>
            curr = self.queue1.pop(0)<br>
            self.queue2.append(curr)<br>
        return self.queue1[0]<br>
<br>
    def empty(self):<br>
#         
<br>
#         :rtype: bool<br>
        
<br>
        return len(self.queue1) + len(self.queue2) == 0
