The guess API is already defined for you.<br>
@param num, your guess<br>
@return -1 if my number is lower, 1 if my number is higher, otherwise return 0<br>
def guess(num):

```python
class Solution(object):
    def guessNumber(self, n):
        """
        :type n: int
        :rtype: int
        """
        # binary search
        begin, end = 1, n
        while begin <= end:
            mid = (begin + end) / 2
            res = guess(mid)
            if res == 0:
                return mid
            elif res > 0:
                begin = mid + 1
            else:
                end = mid - 1
```
